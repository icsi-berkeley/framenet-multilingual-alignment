```python
configs = [
#     ('japanesefn', 'jpn'),
#     ('frenchfn', 'fra'),
    ('spanishfn', 'spa'),
#     ('fnbrasil', 'por'),
#     ('fncopa', 'por')
]
```

```python
from fnalign.loaders import load
from fnalign.models import Alignment
from fnalign.alignment import attribute, muse, wordnet

eng_frm = load("bfn", "en")

for db_name, lang in configs:
    l2_frm = load(db_name, lang)
    alignment = Alignment(db_name, lang)
    alignment.load(eng_frm, l2_frm)
    
    if db_name == "fnbrasil":
        attribute.id_matching(alignment)

    if db_name in ["japanesefn", "spanishfn", "chinesefn", "french"]:
        attribute.name_matching(alignment)

    attribute.fe_matching(alignment)
    
    baseline = alignment.scores[0]["df"]
    fe = alignment.scores[-1]["df"]
    
    for frame, other in alignment.pairs():
        if baseline.loc[frame.name, other.name] > 0 and fe.loc[frame.name, other.name] != 1:
            print(f'{frame.name}, {other.name}')
            en_fes = sorted(fe.name for fe in frame.fes if fe.type == "Core")
            l2_fes = sorted(fe.name for fe in other.fes if fe.type == "Core")
            print(f'\t en: {str(en_fes)}')
            print(f'\t es: {str(l2_fes)}')
            
```
